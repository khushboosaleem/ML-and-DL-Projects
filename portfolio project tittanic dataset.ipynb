### 7.1. Facet grids
A useful approach to explore medium-dimensional data, is by drawing multiple instances of the same plot on different subsets of your dataset. This technique is commonly called as ***â€œlatticeâ€***, or ***â€œtrellisâ€*** plotting, and it is related to the idea of ***â€œsmall multiplesâ€***. To use these features, your data has to be in a Pandas DataFrame.

**Facet grid** forms a matrix of panels defined by row and column by dividing the variables. Due of panels, a single plot looks like multiple plots. It is very helpful to analyze all combinations in two discrete variables.

**FacetGrid** object takes a dataframe as input and the names of the variables that will form the row, column, or hue dimensions of the grid. The variables should be categorical and the data at each level of the variable will be used for a facet along that axis.

The advantage of using Facet is, we can input another variable into the plot. We can make many column facets and align them with the rows of the grid:

#### FacetGrid.map
The main approach for visualizing data on this grid is with the `FacetGrid.map()` method.

***Missing Data***

From the entry total above, there appears to be missing data. A heatmap will help better visualize what features as missing the most information.


```python
# Heatmap for check the missing data
sns.heatmap(df.isnull(), yticklabels = False , cbar = False, cmap = 'tab20c_r')
plt.title('Missing Data: Training Set')
plt.show()
```

The 'Age' variable is missing roughly 20% of its data. This proportion is likely small enough for reasonable replacements using some form of imputation as well (using the knowledge of the other columns to fill in reasonable values).
However, too much data from the 'Cabin' column is missing to do anything useful with it at a basic level. This column may need to be dropped from the data set altogether or change to another feature such as 'Cabin Known: 1 or 0'.  

We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the passengers (imputation).
However, we can be smarter about this and check the average age by passenger class.


```python
plt.figure(figsize = (10,7))
sns.boxplot(x = 'Pclass', y = "Age", data = df, palette = 'GnBu_d').set_title("Age by pessenger class")
plt.show()
```

```python
df["Age"].isnull().sum()
```

```python
# Imputation function
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
    
# Apply the function to the Age column
df['Age'] = df[['Age','Pclass']].apply(impute_age, axis = 1)
```

```python
df['Age'].isnull().sum()
```

```python
df.info()
```

```python
df['Cabin'].isnull().sum()
```

The Cabin column has too many missing values to do anything useful with, so it would be best to remove it from the data frame altogether

```python
# Remove the Cabin feature
df.drop('Cabin', axis = 1, inplace=True)
```

```python
df.info()
```

Since there is only two missing value in Embarked, that observation can just be removed. 

```python
# Remove rows with missing data
df.dropna(inplace=True)
```

```python
df.info()
```

Name and Ticket, passengerId can be removed from the dataset as these features do not provide additional information about passenger's liklihood of survival




```python
# Remove unnecessary columns 
df.drop(['Name', 'Ticket','PassengerId'], axis = 1, inplace=True)
```

```python
df.columns
```

The remaining non-null objebts, Sex and Embarked, will need to be specified as categories for better analysis

```python
# Convert objects to category data type
obj_cat = ['Sex', 'Embarked']

for col_name in obj_cat:
    df[col_name] = df[col_name].astype('category')
```

### **GETTING MODEL READY**

Now that we've explored the data, it is time to get these features 'model ready'. Categorial features will need to be converted into 'dummy variables', otherwise a machine learning algorithm will not be able to take in those features as inputs

```python
# Shape of train data
df.shape
```

```python
# Identify categorical features
df.select_dtypes(['category'])
```

```python
df.select_dtypes('category').columns
```

```python
# Convert categorical features into dummy or indicator fatures
sex = pd.get_dummies(df['Sex'], drop_first = True)
embarked = pd.get_dummies(df['Embarked'], drop_first = True)
```

```python
sex
```

```python
embarked
```

```python
df.head()
```

```python
# Add new dummy columns to data frame
df = pd.concat([df, sex, embarked], axis = 1)
df.head()
```

```python
# Drop unnecessary columns
df.drop(['Sex', 'Embarked'], axis = 1, inplace=True)
print("Train Data Shape: " ,df.shape)
df.head()
```

## ***Apply Machine Learning on the data***

Next, I will feed these features into various classification algorithms to determine the best performance using a simple framework: **Split, Fit, Predict, Score It.**

```python
# Split data to be used in the models
# Create matrix of features
x = df.drop('Survived', axis = 1) # grabs everything else but survived

# Create target Variable
y = df['Survived'] # y is the column trying to predict
```

```python
# Use x ans y variables to split the training data into train and test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .20, random_state = 101)
```

### ***LOGISTIC REGRESSION***

```python
# Fit 
# Import model
from sklearn.linear_model import LogisticRegression

# Create instance of model
logestic_reg = LogisticRegression()

# Pass training data into model
logestic_reg.fit(x_train, y_train)
```

```python
# Predict

y_pred_log_reg = logestic_reg.predict(x_test)
print(y_pred_log_reg)
```

## ***Probablities of each class***

The predict_proba method returns an array where each row sums to 1, as it contains the probablities of the sample belonging to each class. For a binary classification problem, the first column corresponds to the probability of the sample begin in class 0, and the seconds column correspond to the probability of the sample being in class 1. 

```python
prob = logestic_reg.predict_proba(x_test)[:,1]
print(prob.shape)
print(prob)
```

### ***Probability along with the class***

```python
for i in range(len(y_test)):
    print(round(prob[i],3), "-->", y_pred_log_reg[i])
```

### ***Model Evaluation***

```python
# Score It

from sklearn.metrics import classification_report, accuracy_score
print("Classification Model")
# Accuracy
print("--"*40)
logreg_accuracy  = round(accuracy_score(y_test, y_pred_log_reg)*100,2)
print('Accuracy', logreg_accuracy, '%')
```

### **Interpretation**  
**Accuracy**  
82% of the model's predictions are correct.

## ***Decision Tree***

### ***Model Training***

```python
from sklearn.tree import DecisionTreeClassifier

print('Decision Tree Classifier')
# Create instance of model
Dtree = DecisionTreeClassifier()

# Pass training data into model
Dtree.fit(x_train, y_train)
```

## ***Model Evaluation***

### ***Decision Tree Classifier***

```python
from sklearn.metrics import accuracy_score
# Prediction from the model
y_pred_Dtree = Dtree.predict(x_test)
# Score It

print('Decision Tree Classifier')
# Accuracy
print('--'*30)
Dtree_accuracy = round(accuracy_score(y_test, y_pred_Dtree)*100,2)
print('Accuracy', Dtree_accuracy, '%')
```
